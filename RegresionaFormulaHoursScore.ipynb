```python
#Regresiona formula koju dobijate iz Linear Regression modela može se izraziti kao:
#Score=intercept+slope×Hours
#Gde su:
#intercept - presečna tačka (vrednost na y-osi kada je Hours=0),
#slope - koeficijent nagiba, koji predstavlja promenu vrednosti Score u odnosu na promenu vrednosti Hours.
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

data = pd.read_csv('datasets/hours_score.csv')

X = data.loc[:, ['hours']]
y = data['score']

# Inicijalizacija modela linearne regresije
model = LinearRegression()

# Treniranje modela
model.fit(X, y)

# Koeficijenti linearne regresije
intercept = model.intercept_
slope = model.coef_[0]

# Izračunavanje predikcija
predicted_y = model.predict(X)

# Regresiona linija
regression_line = f'Regression line: score = {slope:.2f} * hours + {intercept:.2f}'

# Prikaz rezultata
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, model.predict(X), color='red', label=regression_line)
plt.xlabel('Hours')
plt.ylabel('Score')
plt.legend()
plt.title('Linear Regression')
plt.show()

# Ispis koeficijenata regresije
print(f'Koeficijent nagiba (slope): {slope:.2f}')
print(f'Presečna tačka (intercept): {intercept:.2f}')
```
