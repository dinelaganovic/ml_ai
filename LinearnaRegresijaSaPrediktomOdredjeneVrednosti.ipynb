```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Podaci
data = {
    'Sati_ucenja': [1, 2, 3, 4, 5, 6, 7, 8],
    'Broj_poeni': [30, 45, 51, 57, 60, 65, 70, 71]
}

# Konvertujemo podatke u DataFrame
df = pd.DataFrame(data)

# Razdvajamo podatke na ulaze (X) i izlaz (y)
X = df[['Sati_ucenja']]
y = df['Broj_poeni']

# Inicijalizujemo model linearnu regresiju
model = LinearRegression()

# Treniramo model
model.fit(X, y)

# Izracunavamo predikciju za 10.5 sati ucenja
sati_ucenja_predikcija = [[10.5]]
broj_poeni_predikcija = model.predict(sati_ucenja_predikcija)[0]

# Prikazivanje rezultata
plt.scatter(X, y, color='blue', label='Stvarni podaci')
plt.plot(X, model.predict(X), color='red', label='Linearna regresija')
plt.scatter(sati_ucenja_predikcija, broj_poeni_predikcija, color='green', marker='X', s=100, label='Predikcija za 10.5 sati')
plt.xlabel('Sati ucenja')
plt.ylabel('Broj poena')
plt.legend()
plt.show()

# Izracunavanje greske
predictions = model.predict(X)
mse = mean_squared_error(y, predictions)

print(f'Predikcija za 10.5 sati ucenja: {broj_poeni_predikcija:.2f} poena')
print(f'Srednje-kvadratna greska (MSE): {mse:.2f}')
```
